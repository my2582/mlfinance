```python
import PyPDF2 as pypdf

def findInDict(needle, haystack):
    for key in haystack.keys():
        try:
            value=haystack[key]
        except:
            continue
        if key==needle:
            return value
        if isinstance(value,dict):            
            x=findInDict(needle,value)            
            if x is not None:
                return x
            
pdfobject=open('./MMM_transcript.pdf','rb')
pdf=pypdf.PdfFileReader(pdfobject)
```
